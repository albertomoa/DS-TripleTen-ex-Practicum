#### Memperbaiki Data

Mengubah tipe data kolom `reg_date`untuk memudahkan analisa pengatasan nilai yanghilang pada kolom `churn_date`.

```python
# mengubah tipe data kolom reg_date menjadi date time
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d')
```

```python
# menambahkan kolom baru tahun registrasi user
users['reg_year'] = users['reg_date'].dt.year
```

```python
# menampilkan nilai / isi kolom reg_year
users['reg_year'].unique()
```

Ditemukan hal menarik bahwa semua user mendaftar di tahun yang sama yakni 2018, kemungkinan data ini adalah data pelanggan di tahun 2018, sehingga untuk mengisi kolom yahg hilang pada tabel `churn_date` akan diisi dengan tanggal akhir tahun 2018, yaitu `2018-12-31`.

```python
# mengisi nilai yang hilang pada kolom churn_date
users['churn_date'] = users['churn_date'].fillna('2018-12-31')
```

```python
# mengubah tipe data kolom churn_date menjadi date time
users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%d')
```

Tipe data dan nilai yang hilang berhasil diperbaiki. Selanjutnya akan dilakukan penambahan kolom lama berlangganan pada kolom `sub_duration` dalam satuan bulan.

### Menampilkan Total Pendapatan Setiap Bulan

Untuk menghitung pendapatan dari masing-masing user_id di tiap bulannya, terlebih dahulu ditambahkan kolom baru berikut :
- `ex_call` - berisi kelebihan penggunaan paket telepon
- `ex_message` - berisi kelebihan pengiriman pesan SMS
- `ex-internet`- berisi kelebihan penggunaan internet
- `rev_ex_call` - berisi revenue dari kelebihan penggunaan paket telepon
- `rev_ex_message` - berisi revenue dari kelebihan pengiriman pesan SMS
- `rev_ex_internet` - berisi revenue dari kelebihan penggunaan paket telepon
- `rev_per_user` - berisi pendapatan tap user

```python
# Menambahkan kolom ex_call
new_user_plan_info['ex_call'] = new_user_plan_info['duration'] - new_user_plan_info['minutes_included']

# Menambahkan kolom ex_message
new_user_plan_info['ex_message'] = new_user_plan_info['sum_message'] - new_user_plan_info['messages_included']

# Menambahkan kolom ex_internet
new_user_plan_info['ex_internet'] = new_user_plan_info['mb_used'] - new_user_plan_info['mb_per_month_included']

new_user_plan_info
```

Apabila nilai pada kolom `ex_call`, `ex_message`, dan `ex_internet` bernilai **negatif**, dapat disimpulkan bahwa user bersangkutan tidak melebihi batas penggunaan paket, sehingga nilai **negatif** akan diubah menjadi 0. Apabila bernilai **positif** maka akan dihitung biaya kelebihan setiap paket.

```python
# Membuat fungsi untuk mengubah nilai negatif mejadi 0
def excess(a):
    if a < 0:
        return 0
    else :
        return a
```

```python
# menerapkan fungsi excess pada kolom pendapatan berlebih
new_user_plan_info['ex_call'] = new_user_plan_info['ex_call'].apply(excess)
new_user_plan_info['ex_message'] = new_user_plan_info['ex_message'].apply(excess)
new_user_plan_info['ex_internet'] = new_user_plan_info['ex_internet'].apply(excess)
new_user_plan_info
```

Karena kelebihan penggunaan paket internet dihitung dalam pembulatan Gigabyte, maka kelebihan penggunaan layanan internet akan dibagi dengan `1000` dan dilakukan pembulatan keatas.

```python
# Membagi kolom ex_internet dengan 1024 dan melakukan pembulatan keatas
new_user_plan_info['ex_internet'] = np.ceil(new_user_plan_info['ex_internet'] / 1024)
new_user_plan_info
```

## Menguji Hipotesis Statistik

Pada tahap ini, akan dilakukan uji hipotesis yang sebelumnya telah dirumuskan pada sub-bab tujuan yakni :
1. Rata - rata pendapatan dari pengguna paket telepon `Surf` dan `Ultimate` berbeda.
2. Pendapatan dari pengguna di wilayah NY-NJ adalah yang paling besar.

**1. Menguji Hipotesis rata-rata pendapatan paket telepon `Surf` dan `Ultimate` berbeda**

Untuk melakukan uji hipotesis, sebelumnya akan ditentukan dahulu H0 dan H1 yaitu :

**H0** : Rata-rata pendapatan dari pengguna paket telepon `Surf` sama dengan `Ultimate`

**H1** : Rata-rata pendapatan dari pengguna paket telepon `Surf` dan `Ultimate` berbeda.

Untuk nilai signifikansi statistik, umunya menggunakan `alpha = 5%`, dengan pengujian statistik hipotesis 2 arah antar 2 populasi.

Untuk menguji hipotesis diatas, akan menggunakan informasi dari dataframe `user_plan_surf_revenue` yang berisi rata-rata pendapatan dari pengguna paket `surf` dan `user_plan_ultimate_revenue` yang berisi rata-rata dari pengguna paket `ultimate`.

```python
user_plan_surf_revenue
```

```python
user_plan_ultimate_revenue
```

```python
# menampilkan variansi kedua sampel 
print(f'Variansi pendapatan paket surf : {var_revenue_surf}')
print(f'Variansi pendapatan paket ultimate : {var_revenue_ultimate}')
```

```python
# Menguji hipotesis
alpha = 0.05

results = st.ttest_ind(user_plan_surf_revenue['rev_per_user'], user_plan_ultimate_revenue['rev_per_user'], equal_var=False)

print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print('Menolak hipotesis nol')
else:
    print("Tidak dapat menolak hipotesis nol")
```

Berdasarkan uji hipotesis diatas, didapatkan hasil bahwa tidak dapat menolak hipotesis nol, artinya rata-rata pendapatan dari pengguna paket `Surf` dan `Ultimate` sepanjang tahun 2018 tidak berbeda signifikan.

**2. Pendapatan dari pengguna di wilayah NY-NJ adalah yang paling besar.**

Untuk melakukan uji hipotesis, sebelumnya akan ditentukan dahulu H0 dan H1 yaitu :

**H0** : Rata-rata pendapatan dari pengguna di wilayah NY-NJ sama dengan wilayah lain.

**H1** : Rata-rata pendapatan dari pengguna di wilayah NY-NJ lebih besar dari wilayah lain.

Untuk nilai signifikansi statistik, umunya menggunakan `alpha = 5%`dengan pengujian statistik hipotesis 1 arah.

Untuk melakukan uji hipotesis ini, terlebih dahulu dilakukan agregasi dataframe `users_state` yang berisi wilayah `user_id` dan dataframe `user_plan_revenue` yang berisi informasi rata-rata-pendapatan dari tiap `user_id`.

```python
# mengagregasikan state pengguna dengan dataframe user_plan_revenue
revenue_state = user_plan_revenue.merge(users_state, on='user_id', how='inner')
revenue_state
```

```python
# membuat data frame baru isinya rata-rata revenue berdasarkan pengguna yang berada di wilayah NY-NJ
revenue_NY_NJ_state = revenue_state[revenue_state['state'] == 'NY-NJ-PA']
# menghitung rata-rata pendapatan dari seluruh pengguna wilayah NY-NJ
avg_revenue_NY_NJ_state = revenue_NY_NJ_state['rev_per_user'].mean()

# membuat data frame baru isinya rata-rata revenue berdasarkan state pengguna selain wilayah NY-NJ
avg_revenue_other_state = revenue_state[revenue_state['state'] == 'other_state']
```

```python
# Uji hipotesisnya
alpha = 0.05

results = st.ttest_1samp(avg_revenue_other_state['rev_per_user'], avg_revenue_NY_NJ_state)

print('p-value:', results.pvalue / 2)

if (results.pvalue / 2 < alpha) and (avg_revenue_other_state['rev_per_user'].mean() < avg_revenue_NY_NJ_state):
    print(
        "Menolak hipotesis nol: rata-rata pendapatan wilayah NY-NJ secara signifikan lebih besar dari wilayah lain"
    )
else:
    print(
        "Tidak dapat menolak hipotesis nol: Rata-rata pendapatan wilayah NY-NJ secara signifikan tidak lebih tinggi"
    )
```

Berdasarkan uji statistik diatas, diketahui bahwa rata-rata pendapatan wilayah NY-NJ secara signifikan tidak lebih tinggi dari wilayah lain.
